# Plots of Features
This file contains the visualization of each feature.
We specifically plot the spectrograms based on a dog bark audio file.

```python
%matplotlib inline
import numpy, scipy, matplotlib.pyplot as plt, IPython.display as ipd, sklearn
import librosa, librosa.display
```

### Load the audio clip

```python
x, sr = librosa.load('/Users/moyan/UrbanSound8K/audio/fold1/101415-3-0-2.wav')
ipd.Audio(x, rate=sr)
```

### Wave Format

```python
# plt.figure()
plt.figure(figsize=(10, 5))
librosa.display.waveplot(x, sr=sr)
plt.title('Dog bark in Wave Format')
```

### Spectrogram: MFCC

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5, 4))
mfccs = librosa.feature.mfcc(y=x, sr=sr, n_mfcc=40)
librosa.display.specshow(mfccs, x_axis='time')
plt.colorbar()
plt.title('Dog Bark: MFCC')
plt.tight_layout()
```

### Mel Spectrogram

```python
import numpy as np
plt.figure(figsize=(6, 4))
S = librosa.feature.melspectrogram(y=x, sr=sr, n_mels=128,fmax=8000)
librosa.display.specshow(librosa.power_to_db(S,ref=np.max),y_axis='mel', fmax=8000,x_axis='time', cmap='coolwarm')
plt.colorbar(format='%+2.0f dB')
plt.title('Dog Bark: Mel spectrogram')
```

### Chromogram

```python
C = librosa.feature.chroma_cqt(y=x, sr=sr)
librosa.display.specshow(C, y_axis='chroma', cmap='coolwarm')
plt.colorbar()
plt.title('Sample Chromagram: dog bark')
```

### Spectral Contrast

```python
plt.figure(figsize=(6, 4))
# plt.rcParams['figure.figsize'] = (4, 3)
spectral_contrast = librosa.feature.spectral_contrast(x, sr=sr)
librosa.display.specshow(spectral_contrast, x_axis='time', cmap='coolwarm')
plt.colorbar()
plt.ylabel('Frequency bands')
plt.title('Sample Spectral contrast: dog bark')
```
